## 1. UNIVARIATE ANALYSIS

```python
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt
```

```python
mtcars_data.head()
```

### 1.1 Histogram - 1 Continous Datatypes

```python
%matplotlib inline
plt.hist(x = 'mpg',data=mtcars_data)
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('MPG Distribution')
plt.show()
```

```python
import seaborn as sns
sns.countplot(x='cyl',y=None, data = mtcars_data)
plt.title('CYL Distribution')
plt.show()
```

### 1.3 Boxplot - 1 Continous - Distribution + Outliers

```python
plt.boxplot(x = 'mpg',data=mtcars_data) 
plt.show()
```

```python
mtcars_data['cyl'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.pie(x =mtcars_data['cyl'].value_counts(),labels=['8cyl','4cyl','6cyl'],autopct='%1.0f%%',explode=[0.02,0.02,0.02])
plt.show()
```

## 2. BIVARIATE ANALYSIS

```python
mtcars_data.head()
```

### 2.1 Barplot - 1 Discrete Vs 1 Continous

```python
sns.barplot(x='gear',y='mpg',data = mtcars_data)
plt.title('No.of Gears Vs MPG')
plt.show()
```

### 2.2 ScatterPlot - Cont Vs Cont

```python
sns.scatterplot(x='hp', y='mpg',data = mtcars_data)
plt.title('HP Vs MPG')
plt.show()
```

### 2.3 Boxplot - Distribution + Outliers

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='cyl',y='mpg',data=mtcars_data)
plt.title('Cyl Vs MPG')
plt.show()
```

## 3. MultiVariate Analysis

```python
sns.pairplot(data = mtcars_data)
plt.show()
```

```python
corr_coeff = mtcars_data.corr().round(2)
corr_coeff
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data = corr_coeff,annot=True)
plt.show()
```
